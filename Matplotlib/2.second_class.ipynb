# ðŸ“Š Comprehensive Guide to Matplotlib Plots

Matplotlib is a powerful Python library for data visualization. It allows us to create static, animated, and interactive plots. Below are the most common plots, when to use them, and example code.

## 1. Line Plot (`plt.plot`)

**âœ… Use when:**

* Showing trends over time (time series)
* Visualizing continuous data
* Comparing multiple trends

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, marker='o', color='blue', linestyle='--', label='y = 2x')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot Example")
plt.legend()
plt.show()
```

## 2. Bar Plot (`plt.bar`)

**âœ… Use when:**

* Comparing categories
* Showing discrete data
* Frequency counts

```python
categories = ["A", "B", "C", "D"]
values = [4, 7, 2, 9]

plt.bar(categories, values, color="orange")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Plot Example")
plt.show()
```

## 3. Horizontal Bar Plot (`plt.barh`)

**âœ… Use when:**

* Comparing categories with long labels
* Ranking data

```python
categories = ["Math", "Science", "History", "English"]
scores = [80, 95, 70, 85]

plt.barh(categories, scores, color="green")
plt.xlabel("Scores")
plt.title("Horizontal Bar Plot Example")
plt.show()
```

## 4. Histogram (`plt.hist`)

**âœ… Use when:**

* Understanding distribution of data
* Checking skewness, spread, or outliers

```python
import numpy as np
data = np.random.randn(1000)  # normally distributed data

plt.hist(data, bins=30, color="purple", alpha=0.7)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram Example")
plt.show()
```

## 5. Scatter Plot (`plt.scatter`)

**âœ… Use when:**

* Showing relationships between two variables
* Detecting correlation or clusters
* Spotting outliers

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color="red", marker="x")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Scatter Plot Example")
plt.show()
```

## 6. Pie Chart (`plt.pie`)

**âœ… Use when:**

* Showing proportions or percentages
* Best for categorical distribution

```python
sizes = [30, 20, 25, 25]
labels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=90)
plt.title("Pie Chart Example")
plt.show()
```

## 7. Stack Plot (`plt.stackplot`)

**âœ… Use when:**

* Visualizing cumulative trends over time
* Showing contribution of each category to a total

```python
days = [1, 2, 3, 4, 5]
math = [1, 2, 3, 4, 5]
science = [2, 1, 2, 1, 2]
english = [1, 1, 1, 1, 1]

plt.stackplot(days, math, science, english, labels=["Math", "Science", "English"])
plt.legend(loc="upper left")
plt.title("Stack Plot Example")
plt.show()
```

## 8. Box Plot (`plt.boxplot`)

**âœ… Use when:**

* Checking spread of data (quartiles)
* Detecting outliers
* Comparing multiple groups

```python
data = [7, 8, 5, 6, 9, 7, 8, 6, 5, 10]

plt.boxplot(data)
plt.title("Box Plot Example")
plt.ylabel("Values")
plt.show()
```

## 9. Area Plot (Filled Line Plot)

**âœ… Use when:**

* Showing magnitude over time
* Similar to line plot but emphasizes area under curve

```python
x = np.arange(0, 10, 0.1)
y = np.sin(x)

plt.fill_between(x, y, color="skyblue", alpha=0.5)
plt.plot(x, y, color="blue")
plt.title("Area Plot Example")
plt.show()
```

## 10. Heatmap (via `imshow` or Seaborn)

**âœ… Use when:**

* Visualizing matrices (correlation, confusion matrix, etc.)
* Spotting patterns in 2D data

```python
matrix = np.random.rand(5,5)

plt.imshow(matrix, cmap="viridis", interpolation="nearest")
plt.colorbar()
plt.title("Heatmap Example")
plt.show()
```
