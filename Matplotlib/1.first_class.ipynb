## Introduction to Matplotlib<a name="introduction"></a>

Matplotlib is Python's primary plotting library for creating static, animated, and interactive visualizations.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Sample data for all examples
np.random.seed(42)
x = np.linspace(0, 10, 100)
y = np.sin(x)
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]
```

## Basic Setup and Configuration<a name="setup"></a>

```python
# Basic setup for all plots
plt.figure(figsize=(10, 6))  # Width, height in inches
plt.style.use('seaborn-v0_8')  # Modern style
```

## Line Plots<a name="line-plots"></a>

### When to Use:
- Time series data
- Trend analysis
- Continuous data
- Comparing multiple trends

### Examples:

```python
# Basic line plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='sin(x)', color='blue', linewidth=2, linestyle='-')
plt.plot(x, np.cos(x), label='cos(x)', color='red', linewidth=2, linestyle='--')
plt.title('Sine and Cosine Waves')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Multiple lines with different styles
x = np.linspace(0, 10, 50)
y1 = x * 0.5
y2 = x * 1.0
y3 = x * 1.5

plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='Slow growth', marker='o', markersize=4)
plt.plot(x, y2, label='Medium growth', marker='s', markersize=4)
plt.plot(x, y3, label='Fast growth', marker='^', markersize=4)
plt.title('Growth Comparison')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```

## Bar Plots<a name="bar-plots"></a>

### When to Use:
- Categorical data comparison
- Frequency counts
- Ranking items
- Discrete data visualization

### Examples:

```python
# Vertical bar plot
plt.figure(figsize=(10, 6))
bars = plt.bar(categories, values, color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4'])
plt.title('Sales by Category')
plt.xlabel('Product Categories')
plt.ylabel('Sales (in thousands)')

# Add value labels on bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 1,
             f'{height}', ha='center', va='bottom')

plt.show()
```

```python
# Horizontal bar plot
plt.figure(figsize=(10, 6))
plt.barh(categories, values, color='skyblue')
plt.title('Horizontal Bar Chart')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.show()
```

```python
# Grouped bar plot
categories = ['Q1', 'Q2', 'Q3', 'Q4']
product_a = [20, 35, 30, 35]
product_b = [25, 32, 34, 20]
product_c = [15, 25, 30, 35]

x = np.arange(len(categories))
width = 0.25

plt.figure(figsize=(12, 6))
plt.bar(x - width, product_a, width, label='Product A', alpha=0.8)
plt.bar(x, product_b, width, label='Product B', alpha=0.8)
plt.bar(x + width, product_c, width, label='Product C', alpha=0.8)

plt.xlabel('Quarters')
plt.ylabel('Sales')
plt.title('Quarterly Sales by Product')
plt.xticks(x, categories)
plt.legend()
plt.show()
```

## Scatter Plots<a name="scatter-plots"></a>

### When to Use:
- Correlation analysis
- Outlier detection
- Distribution patterns
- Relationship between two continuous variables

### Examples:

```python
# Basic scatter plot
np.random.seed(42)
x_scatter = np.random.normal(0, 1, 100)
y_scatter = x_scatter * 2 + np.random.normal(0, 1, 100)

plt.figure(figsize=(10, 6))
plt.scatter(x_scatter, y_scatter, alpha=0.6, c='purple', s=50)
plt.title('Scatter Plot Example')
plt.xlabel('X variable')
plt.ylabel('Y variable')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Scatter plot with color coding
sizes = np.random.randint(10, 100, 100)
colors = np.random.randint(1, 4, 100)

plt.figure(figsize=(10, 6))
scatter = plt.scatter(x_scatter, y_scatter, c=colors, s=sizes, alpha=0.6, cmap='viridis')
plt.colorbar(scatter, label='Category')
plt.title('Scatter Plot with Size and Color Encoding')
plt.xlabel('X variable')
plt.ylabel('Y variable')
plt.show()
```

## Histograms<a name="histograms"></a>

### When to Use:
- Distribution analysis
- Frequency distribution
- Identifying patterns in data
- Outlier detection

### Examples:

```python
# Basic histogram
data = np.random.normal(0, 1, 1000)

plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, alpha=0.7, color='lightblue', edgecolor='black')
plt.title('Normal Distribution Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Multiple histograms
data1 = np.random.normal(0, 1, 1000)
data2 = np.random.normal(2, 1.5, 1000)

plt.figure(figsize=(10, 6))
plt.hist(data1, bins=30, alpha=0.5, label='Dataset 1', color='blue')
plt.hist(data2, bins=30, alpha=0.5, label='Dataset 2', color='red')
plt.title('Comparing Two Distributions')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

## Box Plots<a name="box-plots"></a>

### When to Use:
- Statistical summary
- Outlier detection
- Comparing distributions
- Showing median, quartiles, and range

### Examples:

```python
# Basic box plot
data_box = [np.random.normal(0, 1, 100),
           np.random.normal(2, 1.5, 100),
           np.random.normal(-1, 0.5, 100)]

plt.figure(figsize=(10, 6))
plt.boxplot(data_box, tick_labels=['Group A', 'Group B', 'Group C'])
plt.title('Box Plot Comparison')
plt.ylabel('Values')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Horizontal box plot
plt.figure(figsize=(10, 6))
plt.boxplot(data_box, tick_labels=['Group A', 'Group B', 'Group C'], vert=False)
plt.title('Horizontal Box Plot')
plt.xlabel('Values')
plt.show()
```

## Pie Charts<a name="pie-charts"></a>

### When to Use:
- Proportional data
- Percentage distribution
- Part-to-whole relationships
- Simple categorical comparisons

### Examples:

```python
# Basic pie chart
labels = ['Python', 'JavaScript', 'Java', 'C++', 'Other']
sizes = [35, 25, 20, 15, 5]
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#ff99cc']
explode = (0.1, 0, 0, 0, 0)  # explode 1st slice

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', 
        shadow=True, startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures pie is circular
plt.title('Programming Language Popularity')
plt.show()
```

```python
# Donut chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
# Draw a circle at center to create donut
centre_circle = plt.Circle((0,0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Donut Chart Example')
plt.axis('equal')
plt.show()
```

## Heatmaps<a name="heatmaps"></a>

### When to Use:
- Correlation matrices
- Density visualization
- Matrix data representation
- Geographic data

### Examples:

```python
# Correlation heatmap
import seaborn as sns  # For better heatmaps

# Create sample correlation matrix
data = np.random.rand(5, 5)
corr_matrix = np.corrcoef(data)

plt.figure(figsize=(8, 6))
plt.imshow(corr_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar(label='Correlation')
plt.xticks(range(5), ['Var1', 'Var2', 'Var3', 'Var4', 'Var5'])
plt.yticks(range(5), ['Var1', 'Var2', 'Var3', 'Var4', 'Var5'])
plt.title('Correlation Matrix Heatmap')

# Add correlation values
for i in range(5):
    for j in range(5):
        plt.text(j, i, f'{corr_matrix[i, j]:.2f}', 
                ha='center', va='center', color='white' if abs(corr_matrix[i, j]) > 0.5 else 'black')

plt.show()
```

## Subplots<a name="subplots"></a>

### When to Use:
- Multiple related plots
- Dashboard creation
- Comparing different views
- Comprehensive data analysis

### Examples:

```python
# Create 2x2 subplot grid
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Plot 1: Line plot
axes[0, 0].plot(x, np.sin(x), 'b-')
axes[0, 0].set_title('Sine Wave')
axes[0, 0].set_xlabel('X')
axes[0, 0].set_ylabel('sin(X)')

# Plot 2: Scatter plot
axes[0, 1].scatter(x_scatter, y_scatter, alpha=0.6)
axes[0, 1].set_title('Scatter Plot')
axes[0, 1].set_xlabel('X')
axes[0, 1].set_ylabel('Y')

# Plot 3: Bar plot
axes[1, 0].bar(categories, values, color=['red', 'blue', 'green', 'orange'])
axes[1, 0].set_title('Bar Chart')
axes[1, 0].set_xlabel('Categories')
axes[1, 0].set_ylabel('Values')

# Plot 4: Histogram
axes[1, 1].hist(data, bins=30, alpha=0.7, color=['blue']*len(data))
axes[1, 1].set_title('Histogram')
axes[1, 1].set_xlabel('Values')
axes[1, 1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## Customization and Styling<a name="customization"></a>

```python
# Advanced customization example
plt.figure(figsize=(12, 8))

# Create sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Plot with extensive customization
plt.plot(x, y1, label='Sine Wave', color='#2E86AB', linewidth=2.5, 
         linestyle='-', marker='', alpha=0.8)
plt.plot(x, y2, label='Cosine Wave', color='#A23B72', linewidth=2.5, 
         linestyle='--', marker='', alpha=0.8)

# Customize the plot
plt.title('Advanced Customization Example', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('X Axis Label', fontsize=12)
plt.ylabel('Y Axis Label', fontsize=12)

# Customize legend
plt.legend(frameon=True, fancybox=True, shadow=True, framealpha=0.9, 
          fontsize=11, loc='upper right')

# Customize grid
plt.grid(True, alpha=0.3, linestyle='-', linewidth=0.5)

# Customize spines
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(1.5)
ax.spines['bottom'].set_linewidth(1.5)

# Add annotation
plt.annotate('Maximum Point', xy=(np.pi/2, 1), xytext=(np.pi/2 + 1, 0.8),
            arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.2', color='red'),
            fontsize=10)

plt.tight_layout()
plt.show()

plt.savefig('plot1.png')
```

### Common Customization Options:

```python
# Text and fonts
plt.title('Title', fontsize=16, fontweight='bold', fontfamily='serif')
plt.xlabel('X Label', fontsize=14)
plt.ylabel('Y Label', fontsize=14)

# Colors and styles
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4']  # Modern color palette
linestyles = ['-', '--', '-.', ':']
markers = ['o', 's', '^', 'D', 'v']

# Layout and spacing
plt.tight_layout(pad=3.0)  # Adjust spacing
plt.subplots_adjust(wspace=0.3, hspace=0.3)  # For subplots

# Saving figures
plt.savefig('plot.png', dpi=300, bbox_inches='tight', transparent=True)
```

## Quick Reference Guide

| Plot Type | Best For | Key Function | When to Use |
|-----------|----------|--------------|-------------|
| **Line Plot** | Trends over time | `plt.plot()` | Continuous data, time series |
| **Bar Plot** | Categorical comparison | `plt.bar()` | Discrete categories, counts |
| **Scatter Plot** | Relationships | `plt.scatter()` | Correlation, distribution |
| **Histogram** | Distribution | `plt.hist()` | Frequency, data spread |
| **Box Plot** | Statistics | `plt.boxplot()` | Outliers, quartiles |
| **Pie Chart** | Proportions | `plt.pie()` | Percentages, parts of whole |
| **Heatmap** | Matrix data | `plt.imshow()` | Correlations, density |
