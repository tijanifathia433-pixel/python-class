```python
import pandas as pd 
```

```python
read = pd.read_csv("C:/Users/USER/Desktop/python/DATA/Full_Account_Statement.csv")

read
```

```python
read['Debit'] = read['Debit/Credit(₦)'].apply(
    lambda x: float(str(x).replace(",","").replace("₦", "")) if str(x).startswith("-") else 0
)

read['Credit'] = read["Debit/Credit(₦)"].apply(
    lambda x: float(str(x).replace(",","").replace("₦", "")) if str(x).startswith("+") else 0
)

read1 = read.drop(columns=["Debit/Credit(₦)"])

read1
```

```python
read1['Date Time'] = pd.to_datetime(read1['Trans. Time'], format='%Y %b %d %H:%M:%S', errors="coerce")


read1["Date"] = read1['Date Time'].dt.date

read1["Time"] = read1['Date Time'].dt.time
```

```python
read1.head()
```

```python
read1 = read1.drop(columns=[ 'Date Time', 'Trans. Time', 'Value Date','Channel', 'Transaction Reference', 'Counterparty'])
read1
```

```python
read1.fillna({"Descriptions": "Unknown"}, inplace=True)
read1
```

```python
read1.to_csv("../DATA/Clean_Account_Statement.csv")
```

```python
df = pd.read_csv("../DATA/Clean_Account_Statement.csv")

df.iloc[:, 2:5].sum()
```

```python
df['Balance'] = df["Balance(₦)"].apply(
    lambda x: float(str(x).replace(",",""))
)

df['Balance'] = (df["Balance"].astype('float'))
```

```python
df.drop(columns=['Balance(₦)','Unnamed: 0'], inplace=True)

df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns



col = ['Debit', 'Credit','Balance']

data = df[col]

 
data
plt.pie(abs(data.sum()*100), labels=data.columns, explode=(0.1,0,0.1), shadow= True, autopct='%1.2f%%')
plt.title("Credit vs Debit vs Balance")
plt.legend()
plt.show()
```
