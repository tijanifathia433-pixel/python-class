```python
import seaborn as sns 
import matplotlib.pyplot as plt

sns.histplot(arr,bins=10, kde=True)
plt.title("Binomial Distribution (n=20, p=0.7)")
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import random

# Binomial distribution
binom = random.binomial(n=10, p=0.5, size=1000)

# Normal distribution (mean = n*p, std = sqrt(n*p*(1-p)))
mean = 20 * 0.5
std = np.sqrt(20 * 0.5 * 0.5)
normal = random.normal(loc=mean, scale=std, size=1000)

# Plot both
sns.histplot(binom, color="blue", label="Binomial", kde=True)
sns.histplot(normal, color="red", label="Normal Approximation", kde=True)

plt.legend()
plt.title("Binomial vs Normal Distribution")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import random

# Small Î»
poisson_small = random.poisson(lam=5, size=10000)
normal_small = random.normal(loc=5, scale=np.sqrt(5), size=10000)

# Large Î»
poisson_large = random.poisson(lam=50, size=10000)
normal_large = random.normal(loc=50, scale=np.sqrt(50), size=10000)

# Plot small Î»
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.histplot(poisson_small, bins=20, color="blue", stat="density", label="Poisson Î»=5")
sns.histplot(normal_small, bins=20, color="red", stat="density", label="Normal Î¼=5, Ïƒ=âˆš5")
plt.legend()
plt.title("Small Î» (Poor Normal Approximation)")

# Plot large Î»
plt.subplot(1,2,2)
sns.histplot(poisson_large, bins=30, color="blue", stat="density", label="Poisson Î»=50")
sns.histplot(normal_large, bins=30, color="red", stat="density", label="Normal Î¼=50, Ïƒ=âˆš50")
plt.legend()
plt.title("Large Î» (Good Normal Approximation)")

plt.show()
```
